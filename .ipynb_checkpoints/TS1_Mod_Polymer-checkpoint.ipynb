```python
TS1_Current_df_dwell_times = pd.DataFrame(columns =['Cell #', 'Average Hydrogen Production Rate (cc/s) Neutron',
                                                'Average Polymer Production Rate (g/s) Neutron','Average Hydrogen Production Rate (cc/s) Photon',
                                               'Average Polymer Production Rate (g/s) Photon', 'Average Hydrogen Production Rate (cc/s) All Particles',
                                               'Average Polymer Production Rate (g/s) All Particles',
                                                'Max Hydrogen Production Rate (cc/s) Neutron','Max Polymer Production Rate (g/s) Neutron','Average Hydrogen Production Rate (cc/s) Photon',
                                               'Max Polymer Production Rate (g/s) Photon', 'Max Hydrogen Production Rate (cc/s) All Particles',
                                               'Max Polymer Production Rate (g/s) All Particles'])

TS1_Current_df_dwell_times['Cell #'] = TS1_Current_df['Cell #']

TS1_Current_df_dwell_times['Average Hydrogen Production Rate (cc/s) Neutron'] = TS1_Current_df_results['Hydrogen Production Rate (cc/s) Neutron'] * 2.6
TS1_Current_df_dwell_times['Average Polymer Production Rate (g/s) Neutron'] = TS1_Current_df_results['Polymer Production Rate (g/s) Neutron'] * 2.6
TS1_Current_df_dwell_times['Average Hydrogen Production Rate (cc/s) Photon'] = TS1_Current_df_results['Hydrogen Production Rate (cc/s) Photon'] * 2.6
TS1_Current_df_dwell_times['Average Polymer Production Rate (g/s) Photon'] = TS1_Current_df_results['Polymer Production Rate (g/s) Photon'] * 2.6
TS1_Current_df_dwell_times['Average Hydrogen Production Rate (cc/s) All Particles'] = TS1_Current_df_results['Hydrogen Production Rate (cc/s) All Particles'] * 2.6
TS1_Current_df_dwell_times['Average Polymer Production Rate (g/s) All Particles'] = TS1_Current_df_results['Polymer Production Rate (g/s) All Particles'] * 2.6

TS1_Current_df_dwell_times['Max Hydrogen Production Rate (cc/s) Neutron'] = TS1_Current_df_results['Hydrogen Production Rate (cc/s) Neutron'] * 6.8
TS1_Current_df_dwell_times['Max Polymer Production Rate (g/s) Neutron'] = TS1_Current_df_results['Polymer Production Rate (g/s) Neutron'] * 6.8
TS1_Current_df_dwell_times['Max Hydrogen Production Rate (cc/s) Photon'] = TS1_Current_df_results['Hydrogen Production Rate (cc/s) Photon'] * 6.8
TS1_Current_df_dwell_times['Max Polymer Production Rate (g/s) Photon'] = TS1_Current_df_results['Polymer Production Rate (g/s) Photon'] * 6.8
TS1_Current_df_dwell_times['Max Hydrogen Production Rate (cc/s) All Particles'] = TS1_Current_df_results['Hydrogen Production Rate (cc/s) All Particles'] * 6.8
TS1_Current_df_dwell_times['Max Polymer Production Rate (g/s) All Particles'] = TS1_Current_df_results['Polymer Production Rate (g/s) All Particles'] * 6.8

TS1_Current_df_dwell_times
```
