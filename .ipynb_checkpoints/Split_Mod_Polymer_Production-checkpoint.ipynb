## Moles, Radicals per Dose and Mr

```python
#using the G Value in Evans paper of 0.66umol/j
# Mr - molecular weight of polymer - assuming chain length of 20 - (CH2)n 
# - said in Evans to have a parrafin (CnH2n+2) structure but G value uses (CH2)n - Parrafin would make it 282

moles_H = moles_per_MGy(0.66e-6)
print (moles_H, 'mol/g per MGy of Hydrogen')
print (radicals_per_dose(moles_H), 'radicals/g.MGy of Hydrogen')

moles_Pol = moles_per_MGy(0.22e-6)
print (moles_Pol, 'mol/g per MGy of Polymer')
print (radicals_per_dose(moles_Pol), 'radicals/g.MGy of Polymer')

Mr_20 = molecular_weight(20)
print(Mr_20)
```

```python
TS1_Current_Data = {'Cell #': [13, 14, 13000, 13001, 14000, 14001],
                    'CH4 Vol (cc)': [7.82E+01, 7.85E+01, 9.15E+01, 6.71E+01, 9.24E+01, 6.73E+01],
                    'CH4 Mass (g)': [3.45E+01, 3.46E+01, 4.03E+01, 2.96E+01, 4.07E+01, 2.97E+01],
                    'Energy Deposited (MeV/g/s)':[2.54E+12, 2.28E+12, 1.22E+12, 7.71E+11, 1.04E+12, 6.68E+11],
                   }

TS1_Current_df = pd.DataFrame(TS1_Current_Data, columns =['Cell #','CH4 Vol (cc)', 'CH4 Mass (g)', 'Energy Deposited (MeV/g/s)'])
TS1_Current_df
```

#### Adding Dose in MGy/s to DFs

```python
TS1_Current_df['Dose (MGy/s)'] = MeV_per_g_per_sec_to_MGy_per_sec(TS1_Current_df['Energy Deposited (MeV/g/s)'])
TS1_Current_df
#wrie new func to involve mass of methane!!
```

## Hydrogen Gas Production - Current Mod
One mole of hydrogen gas is produced for each atom of hydrogen liberated by irradiation.(Evans)
* im not 100% sure if the volume should be the total volume or the volume per section
* SG is specific gravity - using the 0.43 from Evans paper (but online says its 0.466?)

```python
TS1_Current_Hvol = vol_of_H(moles_H, TS1_Current_df['CH4 Vol (cc)'], TS1_Current_df['Dose (MGy/s)'], 0.43)
```

## Polymer Production - Current Mod

```python
#G value taken from Evans paper - 0.22umol/j for the polymer

TS1_Current_Polymer = rate_of_pol_form(0.22e-6, Mr_20, TS1_Current_df['Dose (MGy/s)'], TS1_Current_df['CH4 Mass (g)'])
```

```python
TS1_Current_Avg_H = TS1_Current_Hvol*2.6
TS1_Current_Avg_Polymer = TS1_Current_Polymer*2.6

TS1_Current_Max_H = TS1_Current_Hvol*6.8
TS1_Current_Max_Polymer = TS1_Current_Polymer*6.8
```

## TS1 Current Mod Results

```python
TS1_Current_df['Hydrogen Production Rate (cc/s)'] = TS1_Current_Hvol
TS1_Current_df['Polymer Production Rate (g/s)'] = TS1_Current_Polymer

TS1_Current_df['Average Volume of Hydrogen (cc)'] = TS1_Current_Avg_H
TS1_Current_df['Average Mass of Polymer (g)'] = TS1_Current_Avg_Polymer

TS1_Current_df['Maximum Volume of Hydrogen (cc)'] = TS1_Current_Max_H
TS1_Current_df['Maximum Mass of Polymer (g)'] = TS1_Current_Max_Polymer

TS1_Current_df
```
